```python

from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
import matplotlib.pyplot as plt

# Open the image
img = Image.open("images/quake.jpg")

# Call Draw method to add 2D graphics on the image
DRAW = ImageDraw.Draw(img)

# Customize font type and font style
CUST_FONT = ImageFont.truetype('fonts/PTSans-BoldItalic.ttf', 100)

TEXT = "display(img)"
text1 = "imshow() \n or \n show()"

# Add text to the image
DRAW.text((650, 200), text=TEXT, fill=(0, 0, 0), font=CUST_FONT)
DRAW.text((700, 800), text=text1, fill=(0, 0, 0), font=CUST_FONT)

# Display the image
plt.imshow(img)
# img.show()

img.save("images/mem1.jpg")
```

```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
import matplotlib.pyplot as plt

# Open the image
img = Image.open("images/assignment.png")
plt.imshow(img)

img1 = Image.open("images/tom.jpg")
plt.imshow(img1)

img2 = Image.open("images/ds.jpg")
plt.imshow(img2)

#resize, first image

img_size = img.size
img1_size = img1.size
img2_size = img2.size


print(img_size,img1_size,img2_size)

img2 = img2.resize(img1.size)
new_image = Image.new('RGB',(2*img1_size[0], img_size[1]+img1_size[1]+50), (250,250,250))
new_image.paste(img,(0,0))
new_image.paste(img1,(0,img_size[1]+50))
new_image.paste(img2,(img1_size[0],img_size[1]+50))
print(new_image.size)
DRAW = ImageDraw.Draw(new_image)

# Customize font type and font style
CUST_FONT = ImageFont.truetype('fonts/PTSans-BoldItalic.ttf', 80)

TEXT = "that \n5th point"

# Add text to the image
DRAW.text((850, 50), text=TEXT, fill=(0, 0, 0), font=CUST_FONT)
new_image.save("images/new_merged_image.jpg","JPEG")
plt.imshow(new_image)
```
